# Daily Forecasting witn Facebook Prophet Library

Prophet is a forecasting library specialized in modelling data with trend, several seasonal patterns and external effects. It is a proven and quick way of forecasting daily data. It is simple, fast and it is available in R and Python. 

The mathematical foundation are additive models, where trend, several seasonal patters and external effects are extracted from the input data. 

The input DataFrame must have two columns: ds (datetime) and y (forecasting variable)

```python
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt
import numpy as np
from fbprophet import Prophet

%matplotlib inline
```

## Build forecsating model

In order to build a forecasting model, the following steps must be followed:
* Split in train and test dataframes
* Instanciate a Prophet class and call fit method on train data
* Make pradiction by calling the predict method from the Prophet instance on test data.

```python
df[-5:]
```

```python
prediction_size = 30
train_df = df[:-prediction_size]
train_df[:5]
```

```python
train_df[-5:]
```

```python
test_df = df[-prediction_size:]
test_df[:5]
```

```python
test_df[-5:]
```

```python
m = Prophet()
m.fit(train_df)
```

```python
future = m.make_future_dataframe(periods=prediction_size)
future[-5:]
```

```python
forecast = m.predict(future)
forecast[-5:]
```

### Forecast vs Actual

```python
m.plot(forecast)
```

###  Model components

* Trend
* Yearly sesasonality
* Weekly seasonality

```python
m.plot_components(forecast)
```

### Forecasting assessment
* Join forecasted and actual values
* Compute performance statistics: MAPE and MAE
* Plot Actual vs Forecast

```python
cmp_df =forecast.set_index('ds')[['yhat', 'yhat_lower', 'yhat_upper']].join(df.set_index('ds'))
cmp_df[:5]
```

```python
def calculate_forecast_errors(df, prediction_size):
    """Calculate MAPE and MAE of the forecast.
    
       Args:
           df: joined dataset with 'y' and 'yhat' columns.
           prediction_size: number of days at the end to predict.
    """
    
    # Now we calculate the values of e_i and p_i according to the formulas given in the article above.
    df['e'] = df['y'] - df['yhat']
    df['p'] = 100 * df['e'] / df['y']
    
    # Recall that we held out the values of the last `prediction_size` days
    # in order to predict them and measure the quality of the model. 
    
    # Now cut out the part of the data which we made our prediction for.
    predicted_part = df[-prediction_size:]
    
    # Define the function that averages absolute error values over the predicted part.
    error_mean = lambda error_name: np.mean(np.abs(predicted_part[error_name]))
    
    # Now we can calculate MAPE and MAE and return the resulting dictionary of errors.
    return {'MAPE': error_mean('p'), 'MAE': error_mean('e')}

for err_name, err_value in calculate_forecast_errors(cmp_df, prediction_size).items():
    print(err_name, err_value)
```

```python
plt.figure(figsize=(12, 9))

plt.plot(cmp_df[-prediction_size:][['y']], 'b', label = 'actual')
plt.plot(cmp_df[-prediction_size:][['yhat']], 'r', label = 'forecast')

plt.title('Call center daily revenue')
plt.ylabel('Reveune')


plt.legend()
```

Using Facebook Prophet is a quick and easy way to obtain forecasts, and it performs well on daily data where ARIMA models may be difficult and time consuming to properly fit. Results are astonishing accurate, considering the effort and time invested to build the model.

This results have been achived with the basic configuration of the model and external variables may be added, sensibility to changes in trend can be set, and some other interesting features can be tunned. In conclusion, this is a very interesing approach to time series forecasting
